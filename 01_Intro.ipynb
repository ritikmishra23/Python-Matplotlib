What is Matplotlib?

Matplotlib is an open-source, low-level plotting library for Python. It allows you to create a wide variety of static, animated, and interactive visualizations such as line plots, bar charts, scatter plots, histograms, and more. It is widely used for data visualization in scientific computing and data analysis.

Pyplot
Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:
import matplotlib.pyplot as plt

```python
#Draw a line in a diagram from position (0,0) to position (5,25):
import matplotlib.pyplot as plt
plt.plot([0,5],[0,25])
plt.show()
```

```python
# Draw two points in the diagram, one at position (1, 3) and one in position (8, 10):
import matplotlib.pyplot as plt
import numpy as np
x_points=np.array([1,8])
y_points=np.array([3,10])
plt.plot(x_points,y_points,'ro') #'ro'=> print red colour circle 

```

Markers
You can use the keyword argument marker to emphasize each point with a specified marker:
example=
Marker	     Description
'o'	         Circle	
'*'	         Star	
'.'	         Point	
','	         Pixel	
'x'	         X	
'X'	         X (filled)	
'+'	         Plus

```python
import matplotlib.pyplot as plt
import numpy as np
x_points=np.array([1,8,10])
y_points=np.array([3,10,5])
plt.plot(x_points,y_points,marker='*')
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r') #marker|line|color
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'solid',linewidth='20.5')
plt.show()
```
